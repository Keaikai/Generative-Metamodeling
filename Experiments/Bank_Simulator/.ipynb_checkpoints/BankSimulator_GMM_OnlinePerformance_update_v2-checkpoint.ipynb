## Simulator Learning with CWGAN

```python
seed=67
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.cuda.manual_seed_all(seed)  # if you are using multi-GPU.
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
    
############################### load data ###############################
df = pd.read_csv("./data/traindata/train_data_rep1.csv",index_col=False,header=0)
testdf = pd.read_csv("./data/onlinetest/test_data_x0.csv",index_col=False,header=0)
################################ CWGAN-GP ################################
    
# Y | X
continuous_vars = ["V11","V12","V13","V14","V15"]
categorical_vars = []
context_vars = ["V1","V2","V3","V4","V5","V6","V7","V8","V9","V10"]
    
# Initialize objects
data_wrapper = wgan.DataWrapper(df, continuous_vars, categorical_vars, 
                            context_vars)
spec = wgan.Specifications(data_wrapper, batch_size=4096, max_epochs=500, critic_lr=1e-3, generator_lr=1e-3,
                            print_every=100, device = "cuda")
generator = wgan.Generator(spec)
critic = wgan.Critic(spec)
    
# train Y | X

y, context = data_wrapper.preprocess(df)
wgan.train(generator, critic, y, context, spec)
```

```python
run=100
OnlineTime_CWGAN=np.zeros((run,1))
for runi in np.arange(0,run):
    random.seed(runi)
    np.random.seed(runi)
    torch.manual_seed(runi)
    torch.cuda.manual_seed(runi)
    torch.cuda.manual_seed_all(runi)  # if you are using multi-GPU.
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    # simulate data with conditional WGANs
    time_start = time.time() #开始计时
    testdf_generated = data_wrapper.apply_generator(generator, testdf)
    time_end = time.time() #结束计时 
    OnlineTime_CWGAN[runi,0]= time_end - time_start
    testdf_generated.to_csv("./data/onlinetest/testdata_CWGAN_online/testdata_CWGAN_online_rep" + '{}.csv'.format(runi+1),index=0)
OnlineTime_CWGAN =pd.DataFrame(OnlineTime_CWGAN)
OnlineTime_CWGAN.to_csv('./data/OnlineTime_CWGAN.csv',index=0)    
```

## PLot Marginal Histograms (Conditional)

```python
runi=0
testdf = pd.read_csv("./data/onlinetest/test_data_x0.csv".format(runi+1),index_col=False,header=0)
testdf_Y=testdf.iloc[:,10:15]
#testdf_Y.columns=['average_system_time_normal','renege_rate_normal','average_system_time_priority','renege_rate_priority','max_overtime']
testdf_Y.columns=['$Y_1$','$Y_2$','$Y_3$','$Y_4$','$Y_5$']
testdf_QRNNGMM=pd.read_csv('./data/onlinetest/testdata_QRNNGMM_online/testdata_QRNNGMM_online_rep' + '{}.csv'.format(runi+1),index_col=0,header=0)
testdf_QRNNGMM_Y=testdf_QRNNGMM.iloc[:,10:15]
#testdf_QRNNGMM_Y.columns=['average_system_time_normal','renege_rate_normal','average_system_time_priority','renege_rate_priority','max_overtime']
testdf_QRNNGMM_Y.columns=['$Y_1$','$Y_2$','$Y_3$','$Y_4$','$Y_5$']
testdf_CWGAN= pd.read_csv('./data/onlinetest/testdata_CWGAN_online/testdata_CWGAN_online_rep' + '{}.csv'.format(runi+1),index_col=None,header=0)
testdf_CWGAN_Y=testdf_CWGAN.iloc[:,10:15]
#testdf_CWGAN_Y.columns=['average_system_time_normal','renege_rate_normal','average_system_time_priority','renege_rate_priority','max_overtime']
testdf_CWGAN_Y.columns=['$Y_1$','$Y_2$','$Y_3$','$Y_4$','$Y_5$']
```

```python
#histogram=dict(variables=['average_system_time_normal','renege_rate_normal','average_system_time_priority','renege_rate_priority','max_overtime'], nrow=1, ncol=5)
histogram=dict(variables=['$Y_1$','$Y_2$','$Y_3$','$Y_4$','$Y_5$'], nrow=1, ncol=5)
figsize=10
save=True
path="./plot/"
```

```python
# histogram marginals
if histogram and len(histogram["variables"]) > 0:
    fig2, axarr2 = plt.subplots(histogram["nrow"], histogram["ncol"],
                                figsize=(42.5, 8.5))
    v = 0
    for i in range(histogram["nrow"]):
        for j in range(histogram["ncol"]):
            plot_var, v = histogram["variables"][v], v+1
            axarr2[j].hist([testdf_Y[plot_var], testdf_QRNNGMM_Y[plot_var], testdf_CWGAN_Y[plot_var]], bins=8, density=False,
                            histtype='bar', label=["Real", r'QRGMM$^{+}$',"CWGAN"], color=["#0DE4B8","#5C00FF", "#FF0000"])
            axarr2[j].legend(prop={"size": 20})
            axarr2[j].ticklabel_format(style='sci', scilimits=(0,0)) 
            axarr2[j].xaxis.get_offset_text().set(size=25)
            axarr2[j].tick_params(labelsize=25)
            axarr2[j].yaxis.get_offset_text().set(size=25)
            axarr2[j].set_title(plot_var,fontsize=30)
    if save:
        fig2.savefig(path+'conditional_marginaldist_hist_update.pdf',bbox_inches='tight')
    else:
        fig2.show()
```
