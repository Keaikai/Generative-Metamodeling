```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
run = 100
p, d = 10, 5
Y = slice(p, p + d)

DATA_DIR = "./data"
OUT_DIR  = "./results"
os.makedirs(OUT_DIR, exist_ok=True)

PATHS = {
    "Truth":     DATA_DIR + "/testdata/test_data_rep{rep}.csv",
    "QRGMM+":    DATA_DIR + "/testdata_QRNNGMM/testdata_QRNNGMM_rep{rep}.csv",
    "CWGAN":     DATA_DIR + "/testdata_CWGAN/testdata_CWGAN_rep{rep}.csv",
    "Diffusion": DATA_DIR + "/testdata_Diffusion/testdata_Diffusion_rep{rep}.csv",
    "RectFlow":  DATA_DIR + "/testdata_RectFlow/testdata_RectFlow_rep{rep}.csv",
}
METHODS = ["Truth", "QRGMM+", "CWGAN", "Diffusion", "RectFlow"]

# only QRNNGMM may have an extra index column
INDEX_COL = {"QRGMM+": 0}


def read_df(path, index_col=None):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Missing file: {path}")
    return pd.read_csv(path, header=0, index_col=index_col)


def summarize_over_reps(arr_run_by_d):
    """arr shape: (run, d) -> returns (d,), (d,)"""
    return arr_run_by_d.mean(axis=0), arr_run_by_d.std(axis=0, ddof=1)


def make_columns(d=5):
    # 10 columns: Y1_mean, Y1_sd, ..., Y5_mean, Y5_sd
    cols = []
    for j in range(1, d + 1):
        cols += [(f"Y{j}", "Mean"), (f"Y{j}", "SD")]
    return pd.MultiIndex.from_tuples(cols)


# ================= main computation =================
# store per-rep stats: (run, d)
rep_mean = {m: np.zeros((run, d)) for m in METHODS}  # per rep: sample mean of Y
rep_sd   = {m: np.zeros((run, d)) for m in METHODS}  # per rep: sample sd of Y

for rep in range(1, run + 1):
    for m in METHODS:
        df = read_df(PATHS[m].format(rep=rep), index_col=INDEX_COL.get(m, None))
        rep_mean[m][rep - 1] = df.iloc[:, Y].mean().to_numpy()
        rep_sd[m][rep - 1]   = df.iloc[:, Y].std(ddof=1).to_numpy()

# Build two readable tables:
# 1) Sample Mean block: mean/sd over reps of (sample means)
# 2) Sample SD block: mean/sd over reps of (sample sds)
cols = make_columns(d)

table_sample_mean = pd.DataFrame(index=METHODS, columns=cols, dtype=float)
table_sample_sd   = pd.DataFrame(index=METHODS, columns=cols, dtype=float)

for m in METHODS:
    mu_mean, sd_mean = summarize_over_reps(rep_mean[m])  # each (d,)
    mu_sd,   sd_sd   = summarize_over_reps(rep_sd[m])    # each (d,)

    # fill 10 columns
    row_mean = np.column_stack([mu_mean, sd_mean]).reshape(-1)  # (10,)
    row_sd   = np.column_stack([mu_sd,   sd_sd]).reshape(-1)    # (10,)

    table_sample_mean.loc[m, :] = row_mean
    table_sample_sd.loc[m, :]   = row_sd

# nicer printing
pd.set_option("display.width", 200)
pd.set_option("display.max_columns", 50)

print("\n===================== Sample Mean =====================")
print(table_sample_mean.round(4).to_string())

print("\n===================== Sample SD =======================")
print(table_sample_sd.round(4).to_string())

# save for later plotting
table_sample_mean.to_csv(os.path.join(OUT_DIR, "bank_table_sample_mean.csv"))
table_sample_sd.to_csv(os.path.join(OUT_DIR, "bank_table_sample_sd.csv"))
print(f"\n[Saved] {OUT_DIR}/bank_table_sample_mean.csv")
print(f"[Saved] {OUT_DIR}/bank_table_sample_sd.csv")
```

```python
# -----------------------------
# Load online generation times
# -----------------------------
time_QRNNGMM = pd.read_csv("./data/onlinetest/onlinetime_QRGMM.csv").iloc[:, -1].values
time_CWGAN   = pd.read_csv("./data/onlinetest/onlinetime_CWGAN.csv").iloc[:, 0].values
time_Diff    = pd.read_csv("./data/onlinetest/onlinetime_Diffusion.csv").iloc[:, 0].values
time_Rect    = pd.read_csv("./data/onlinetest/onlinetime_RectFlow.csv").iloc[:, 0].values

# -----------------------------
# Summary table (mean / std)
# -----------------------------
summary_time = pd.DataFrame({
    "mean(time)": [
        time_QRNNGMM.mean(),
        time_CWGAN.mean(),
        time_Diff.mean(),
        time_Rect.mean()
    ],
    "std(time)": [
        time_QRNNGMM.std(ddof=1),
        time_CWGAN.std(ddof=1),
        time_Diff.std(ddof=1),
        time_Rect.std(ddof=1)
    ]
}, index=["QRNNGMM", "CWGAN", "Diffusion", "RectFlow"])

print("\nOnline generation time summary (seconds):\n")
print(summary_time.round(4))
```

## PLot Marginal Histograms (Conditional)

```python
plt.rcParams['hatch.linewidth'] = 1.5
# plt.rcParams['hatch.color'] = 'k'

runi = 1

# =========================
# Load truth + model outputs (online conditional)
# =========================
testdf = pd.read_csv("./data/onlinetest/test_data_x0.csv", index_col=False, header=0)
testdf_Y = testdf.iloc[:, 10:15].copy()
testdf_Y.columns = [r"$Y_1$", r"$Y_2$", r"$Y_3$", r"$Y_4$", r"$Y_5$"]

testdf_QRNNGMM = pd.read_csv(
    f"./data/onlinetest/testdata_QRNNGMM_online/testdata_QRNNGMM_online_rep{runi+1}.csv",
    index_col=0, header=0
)
testdf_QRNNGMM_Y = testdf_QRNNGMM.iloc[:, 10:15].copy()
testdf_QRNNGMM_Y.columns = testdf_Y.columns

testdf_CWGAN = pd.read_csv(
    f"./data/onlinetest/testdata_CWGAN_online/testdata_CWGAN_online_rep{runi+1}.csv",
    index_col=None, header=0
)
testdf_CWGAN_Y = testdf_CWGAN.iloc[:, 10:15].copy()
testdf_CWGAN_Y.columns = testdf_Y.columns

# --- add Diffusion ---
testdf_Diffusion = pd.read_csv(
    f"./data/onlinetest/testdata_Diffusion_online/testdata_Diffusion_online_rep{runi+1}.csv",
    index_col=None, header=0
)
testdf_Diffusion_Y = testdf_Diffusion.iloc[:, 10:15].copy()
testdf_Diffusion_Y.columns = testdf_Y.columns

# --- add RectFlow ---
testdf_RectFlow = pd.read_csv(
    f"./data/onlinetest/testdata_RectFlow_online/testdata_RectFlow_online_rep{runi+1}.csv",
    index_col=None, header=0
)
testdf_RectFlow_Y = testdf_RectFlow.iloc[:, 10:15].copy()
testdf_RectFlow_Y.columns = testdf_Y.columns

# =========================
# Plot config
# =========================
histogram = dict(variables=list(testdf_Y.columns), nrow=1, ncol=5)
save = True
path = "./plot/"
os.makedirs(path, exist_ok=True)

labels  = ["Real", r"QRGMM$^{+}$", "CWGAN", "Diffusion", "RectFlow"]
colors  = ["green", "blue", "red", "tab:orange", "tab:purple"]
hatches = ["", "..", "\\\\", "xx", "///"]   # aligned to WD style
linew = 0.8

# =========================
# Histogram marginals
# =========================
if histogram and len(histogram["variables"]) > 0:
    fig2, axarr2 = plt.subplots(histogram["nrow"], histogram["ncol"], figsize=(42.5, 8.5))

    v = 0
    for i in range(histogram["nrow"]):
        for j in range(histogram["ncol"]):
            plot_var = histogram["variables"][v]
            v += 1

            data_list = [
                testdf_Y[plot_var].to_numpy(),
                testdf_QRNNGMM_Y[plot_var].to_numpy(),
                testdf_CWGAN_Y[plot_var].to_numpy(),
                testdf_Diffusion_Y[plot_var].to_numpy(),
                testdf_RectFlow_Y[plot_var].to_numpy(),
            ]

            # hist returns (n, bins, patches); patches is list of BarContainer (one per dataset)
            n, bins, patch_containers = axarr2[j].hist(
                data_list,
                bins=5,
                density=False,
                histtype="bar",
                label=labels,
                color=colors,
                edgecolor="w",
                linewidth=linew,
                alpha=0.8
            )

            # Apply per-dataset hatch to each BarContainer
            for k, bar_container in enumerate(patch_containers):
                for rect in bar_container.patches:
                    rect.set_hatch(hatches[k])

            if j == histogram["ncol"] - 1:
                axarr2[j].legend(prop={"size": 28})
            axarr2[j].ticklabel_format(style="sci", scilimits=(0, 0))
            axarr2[j].xaxis.get_offset_text().set(size=35)
            axarr2[j].yaxis.get_offset_text().set(size=35)
            axarr2[j].tick_params(labelsize=35)
            axarr2[j].set_title(plot_var, fontsize=40)

    if save:
        fig2.savefig(os.path.join(path, "conditional_marginaldist_hist.pdf"),
                     bbox_inches="tight", dpi=300)
    else:
        plt.show()
```
