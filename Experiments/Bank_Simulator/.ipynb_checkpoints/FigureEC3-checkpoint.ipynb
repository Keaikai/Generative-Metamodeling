```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import random
import ot
import os
```

```python
run=100
```

```python
def wd_distance(real, gen):
    n = real.shape[0]
    a = np.ones(n)/n
    d_gen = ot.emd2(a, a, M=ot.dist(real.to_numpy(),
                              gen.to_numpy(),
                              metric='euclidean'),numItermax=2000000)
    return d_gen
```

```python
p = 10
d = 5

wd_CWGAN      = np.zeros((run, 1))
wd_QRNNGMM    = np.zeros((run, 1))
wd_Diffusion  = np.zeros((run, 1))
wd_RectFlow   = np.zeros((run, 1))


for runi in np.arange(0, run):
    df = pd.read_csv(f"./data/traindata/train_data_rep{runi+1}.csv", index_col=False, header=0)
    testdf = pd.read_csv(f"./data/testdata/test_data_rep{runi+1}.csv", index_col=False, header=0)

    testdf_CWGAN = pd.read_csv(f"./data/testdata_CWGAN/testdata_CWGAN_rep{runi+1}.csv", index_col=None, header=0)
    testdf_QRNNGMM = pd.read_csv(f"./data/testdata_QRNNGMM/testdata_QRNNGMM_rep{runi+1}.csv", index_col=0, header=0)
    testdf_Diffusion = pd.read_csv(f"./data/testdata_Diffusion/testdata_Diffusion_rep{runi+1}.csv", index_col=None, header=0)
    testdf_RectFlow  = pd.read_csv(f"./data/testdata_RectFlow/testdata_RectFlow_rep{runi+1}.csv", index_col=None, header=0)

    # ---------- WD (truth vs generated) ----------
    Y_true = testdf.iloc[:, 10:15]
    wd_CWGAN[runi]     = wd_distance(Y_true, testdf_CWGAN.iloc[:, 10:15])
    wd_QRNNGMM[runi]   = wd_distance(Y_true, testdf_QRNNGMM.iloc[:, 10:15])
    wd_Diffusion[runi] = wd_distance(Y_true, testdf_Diffusion.iloc[:, 10:15])
    wd_RectFlow[runi]  = wd_distance(Y_true, testdf_RectFlow.iloc[:, 10:15])
```

```python
os.makedirs("./results", exist_ok=True)

wd_df = pd.DataFrame({
    "rep": np.arange(1, run + 1),
    "CWGAN": wd_CWGAN.flatten(),
    "QRNNGMM": wd_QRNNGMM.flatten(),
    "Diffusion": wd_Diffusion.flatten(),
    "RectFlow": wd_RectFlow.flatten()
})

wd_df.to_csv("./results/bank_wd.csv", index=False)
print("WD saved to:")
print("  ./results/bank_wd.csv")
```

## Plot WD

```python
# =======================
# 1) Read saved WD results
# =======================
wd_path = "./results/bank_wd.csv"
wd_df = pd.read_csv(wd_path)

# ---- make sure 1D arrays ----
WD_CWGAN     = wd_df["CWGAN"].to_numpy().reshape(-1)
WD_QRNNGMM   = wd_df["QRNNGMM"].to_numpy().reshape(-1)
WD_Diffusion = wd_df["Diffusion"].to_numpy().reshape(-1)
WD_RectFlow  = wd_df["RectFlow"].to_numpy().reshape(-1)

# (optional) drop NaN/inf just in case
def clean(x):
    x = np.asarray(x, dtype=float).reshape(-1)
    return x[np.isfinite(x)]

WD_CWGAN     = clean(WD_CWGAN)
WD_QRNNGMM   = clean(WD_QRNNGMM)
WD_Diffusion = clean(WD_Diffusion)
WD_RectFlow  = clean(WD_RectFlow)
```

```python
plt.rcParams['hatch.linewidth'] = 1.5
# plt.rcParams['hatch.color'] = 'k'

fig, ax = plt.subplots(1, 1, figsize=(13, 6.5))

# bins: keep your original setting
bins = np.arange(0.5, 6, 0.08)

# ========================= WD hist (style aligned to template) =========================
ax.hist(WD_CWGAN, bins=bins, density=False,
        color='red', alpha=0.6, label='CWGAN',
        edgecolor='w', hatch='\\\\', linewidth=0.8)

ax.hist(WD_Diffusion, bins=bins, density=False,
        color='tab:orange', alpha=0.6, label='Diffusion',
        edgecolor='w', hatch='xx', linewidth=0.8)

ax.hist(WD_RectFlow, bins=bins, density=False,
        color='tab:purple', alpha=0.6, label='RectFlow',
        edgecolor='w', hatch='///', linewidth=0.8)

ax.hist(WD_QRNNGMM, bins=bins, density=False,
        color='blue', alpha=0.6, label=r'QRGMM$^{+}$',
        edgecolor='w', hatch='..', linewidth=0.8)

ax.legend(fontsize=15)
ax.tick_params(labelsize=18)
#ax.set_title("WD (with table)", fontsize=18)

# ========================= table (style aligned to template) =========================
col_labels = ['Mean', 'Std']
row_labels = ['CWGAN', 'Diffusion', 'RectFlow', r'QRGMM$^{+}$']
table_vals = [
    [f"{WD_CWGAN.mean():.2f}",     f"{WD_CWGAN.std(ddof=1):.2f}"],
    [f"{WD_Diffusion.mean():.2f}", f"{WD_Diffusion.std(ddof=1):.2f}"],
    [f"{WD_RectFlow.mean():.2f}",  f"{WD_RectFlow.std(ddof=1):.2f}"],
    [f"{WD_QRNNGMM.mean():.2f}",   f"{WD_QRNNGMM.std(ddof=1):.2f}"],
]

my_table = ax.table(
    cellText=table_vals,
    colWidths=[0.1] * 2,
    rowLabels=row_labels,
    colLabels=col_labels,
    edges='closed',
    loc='center right'
)
my_table.auto_set_font_size(False)
my_table.set_fontsize(15)
my_table.scale(2, 2)

os.makedirs("./plot", exist_ok=True)
fig.savefig("./plot/BankSimulator_WD.png", dpi=300, bbox_inches='tight')
plt.show()
```
