```python
import numpy as np
import pandas as pd
```

```python
run=100
# -------------------------
# allocate (Y1)
# -------------------------
mean_test1_1     = np.zeros((run, 1))
mean_QRGMM1_1    = np.zeros((run, 1))
mean_CWGAN1_1    = np.zeros((run, 1))
mean_Diff1_1     = np.zeros((run, 1))
mean_Rect1_1     = np.zeros((run, 1))

# -------------------------
# allocate (Y2)
# -------------------------
mean_test1_2     = np.zeros((run, 1))
mean_QRGMM1_2    = np.zeros((run, 1))
mean_CWGAN1_2    = np.zeros((run, 1))
mean_Diff1_2     = np.zeros((run, 1))
mean_Rect1_2     = np.zeros((run, 1))

# -------------------------
# loop
# -------------------------
for runi in range(run):

    # ===== Y1 =====
    testdf1_1 = pd.read_csv(f"./data/testdf1_1/testdf1_1_{runi+1}.csv", header=None)
    qrgmm1_1  = pd.read_csv(f"./data/QRGMMoutput1_1/QRGMMoutput1_1_{runi+1}.csv", header=None)
    cwgan1_1  = pd.read_csv(f"./data/CWGANoutput1_1/CWGANoutput1_1_{runi+1}.csv")
    diff1_1   = pd.read_csv(f"./data/Diffusionoutput1_1/Diffusionoutput1_1_{runi+1}.csv")
    rect1_1   = pd.read_csv(f"./data/RectFlowoutput1_1/RectFlowoutput1_1_{runi+1}.csv")

    mean_test1_1[runi]  = testdf1_1.iloc[:, 4].mean()
    mean_QRGMM1_1[runi] = qrgmm1_1.iloc[:, 0].mean()
    mean_CWGAN1_1[runi] = cwgan1_1.iloc[:, 4].mean()
    mean_Diff1_1[runi]  = diff1_1.iloc[:, 4].mean()
    mean_Rect1_1[runi]  = rect1_1.iloc[:, 4].mean()

    # ===== Y2 =====
    testdf1_2 = pd.read_csv(f"./data/testdf1_2/testdf1_2_{runi+1}.csv", header=None)
    qrgmm1_2  = pd.read_csv(f"./data/QRGMMoutput1_2/QRGMMoutput1_2_{runi+1}.csv", header=None)
    cwgan1_2  = pd.read_csv(f"./data/CWGANoutput1_2/CWGANoutput1_2_{runi+1}.csv")
    diff1_2   = pd.read_csv(f"./data/Diffusionoutput1_2/Diffusionoutput1_2_{runi+1}.csv")
    rect1_2   = pd.read_csv(f"./data/RectFlowoutput1_2/RectFlowoutput1_2_{runi+1}.csv")

    mean_test1_2[runi]  = testdf1_2.iloc[:, 4].mean()
    mean_QRGMM1_2[runi] = qrgmm1_2.iloc[:, 0].mean()
    mean_CWGAN1_2[runi] = cwgan1_2.iloc[:, 4].mean()
    mean_Diff1_2[runi]  = diff1_2.iloc[:, 4].mean()
    mean_Rect1_2[runi]  = rect1_2.iloc[:, 4].mean()

pred_LR_1=pd.read_csv("./data/LinearRegression/y1_pred.csv")
pred_LR_2=pd.read_csv("./data/LinearRegression/y2_pred.csv")
pred_LR_1=pred_LR_1.to_numpy()
pred_LR_2=pred_LR_2.to_numpy()
```

```python
summary_Y1 = pd.DataFrame({
    "mean(mean)": [
        mean_test1_1.mean(),
        mean_QRGMM1_1.mean(),
        mean_CWGAN1_1.mean(),
        mean_Diff1_1.mean(),
        mean_Rect1_1.mean(),
        pred_LR_1.mean()
    ],
    "std(mean)": [
        mean_test1_1.std(ddof=1),
        mean_QRGMM1_1.std(ddof=1),
        mean_CWGAN1_1.std(ddof=1),
        mean_Diff1_1.std(ddof=1),
        mean_Rect1_1.std(ddof=1),
        pred_LR_1.std(ddof=1)
    ]
}, index=[
    "Truth",
    "QRGMM",
    "CWGAN",
    "Diffusion",
    "RectFlow",
    "LinearReg"
])

print("\nSummary (Y1):\n")
print(summary_Y1.round(4))
```

```python
summary_Y2 = pd.DataFrame({
    "mean(mean)": [
        mean_test1_2.mean(),
        mean_QRGMM1_2.mean(),
        mean_CWGAN1_2.mean(),
        mean_Diff1_2.mean(),
        mean_Rect1_2.mean(),
        pred_LR_2.mean()
    ],
    "std(mean)": [
        mean_test1_2.std(ddof=1),
        mean_QRGMM1_2.std(ddof=1),
        mean_CWGAN1_2.std(ddof=1),
        mean_Diff1_2.std(ddof=1),
        mean_Rect1_2.std(ddof=1),
        pred_LR_2.std(ddof=1)
    ]
}, index=[
    "Truth",
    "QRGMM",
    "CWGAN",
    "Diffusion",
    "RectFlow",
    "LinearReg"
])

print("\nSummary (Y2):\n")
print(summary_Y2.round(4))
```

```python
correct_prob_sim=sum(mean_test1_1<mean_test1_2)/run
correct_prob_QRGMM=sum(mean_QRGMM1_1<mean_QRGMM1_2)/run
correct_prob_CWGAN=sum(mean_CWGAN1_1<mean_CWGAN1_2)/run
correct_prob_LR=sum(pred_LR_1<pred_LR_2)/run
```

```python
# -------------------------
# Percentage of Correct Ranking
# -------------------------
correct_prob_sim       = float(np.sum(mean_test1_1 < mean_test1_2) / run)
correct_prob_QRGMM     = float(np.sum(mean_QRGMM1_1 < mean_QRGMM1_2) / run)
correct_prob_CWGAN     = float(np.sum(mean_CWGAN1_1 < mean_CWGAN1_2) / run)
correct_prob_Diffusion = float(np.sum(mean_Diff1_1  < mean_Diff1_2)  / run)
correct_prob_RectFlow  = float(np.sum(mean_Rect1_1  < mean_Rect1_2)  / run)
correct_prob_LR        = float(sum(pred_LR_1<pred_LR_2)/run)

print("\nPercentage of Correct Ranking:\n")
print(f"Truth (Simulator): {correct_prob_sim:.4f}")
print(f"QRGMM:             {correct_prob_QRGMM:.4f}")
print(f"CWGAN:             {correct_prob_CWGAN:.4f}")
print(f"Diffusion:         {correct_prob_Diffusion:.4f}")
print(f"RectFlow:          {correct_prob_RectFlow:.4f}")
print(f"LinearReg:         {correct_prob_LR:.4f}")
```
