```python
import numpy as np
import pandas as pd
```

```python
run = 100

# ---------- allocate ----------
mean_test       = np.zeros(run)
mean_QRGMM      = np.zeros(run)
mean_CWGAN      = np.zeros(run)
mean_Diffusion  = np.zeros(run)
mean_RectFlow   = np.zeros(run)

std_test        = np.zeros(run)
std_QRGMM       = np.zeros(run)
std_CWGAN       = np.zeros(run)
std_Diffusion   = np.zeros(run)
std_RectFlow    = np.zeros(run)

# ---------- loop ----------
for runi in range(run):
    testdf = pd.read_csv(f"./data/testdata/testdata_{runi+1}.csv")
    test_gen_QRGMM = pd.read_csv(f"./data/testdata_QRGMM/testdata_QRGMM_{runi+1}.csv")
    test_gen_CWGAN = pd.read_csv(f"./data/testdata_CWGAN/testdata_CWGAN_{runi+1}.csv")
    test_gen_Diffusion = pd.read_csv(f"./data/testdata_Diffusion/testdata_Diffusion_{runi+1}.csv")
    test_gen_RectFlow  = pd.read_csv(f"./data/testdata_RectFlow/testdata_RectFlow_{runi+1}.csv")

    # ----- mean -----
    mean_test[runi]      = testdf.iloc[:, 3].mean()
    mean_QRGMM[runi]     = test_gen_QRGMM.iloc[:, 0].mean()
    mean_CWGAN[runi]     = test_gen_CWGAN.iloc[:, 3].mean()
    mean_Diffusion[runi] = test_gen_Diffusion.iloc[:, 3].mean()
    mean_RectFlow[runi]  = test_gen_RectFlow.iloc[:, 3].mean()

    # ----- std -----
    std_test[runi]       = testdf.iloc[:, 3].std(ddof=1)
    std_QRGMM[runi]      = test_gen_QRGMM.iloc[:, 0].std(ddof=1)
    std_CWGAN[runi]      = test_gen_CWGAN.iloc[:, 3].std(ddof=1)
    std_Diffusion[runi]  = test_gen_Diffusion.iloc[:, 3].std(ddof=1)
    std_RectFlow[runi]   = test_gen_RectFlow.iloc[:, 3].std(ddof=1)
```

```python
summary = pd.DataFrame({
    "mean(mean)": [
        mean_test.mean(),
        mean_QRGMM.mean(),
        mean_CWGAN.mean(),
        mean_Diffusion.mean(),
        mean_RectFlow.mean()
    ],
    "std(mean)": [
        mean_test.std(ddof=1),
        mean_QRGMM.std(ddof=1),
        mean_CWGAN.std(ddof=1),
        mean_Diffusion.std(ddof=1),
        mean_RectFlow.std(ddof=1)
    ],
    "mean(std)": [
        std_test.mean(),
        std_QRGMM.mean(),
        std_CWGAN.mean(),
        std_Diffusion.mean(),
        std_RectFlow.mean()
    ],
    "std(std)": [
        std_test.std(ddof=1),
        std_QRGMM.std(ddof=1),
        std_CWGAN.std(ddof=1),
        std_Diffusion.std(ddof=1),
        std_RectFlow.std(ddof=1)
    ]
}, index=["Truth", "QRGMM", "CWGAN", "Diffusion", "RectFlow"])

print("\nSummary statistics:\n")
print(summary.round(4))
```
