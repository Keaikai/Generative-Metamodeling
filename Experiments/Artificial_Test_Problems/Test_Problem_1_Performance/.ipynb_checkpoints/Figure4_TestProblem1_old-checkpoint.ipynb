```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import wasserstein_distance
```

```python
run=100
WD_QRGMM=np.zeros(run)
WD_CWGAN=np.zeros(run)

KS_QRGMM=np.zeros(run)
KS_CWGAN=np.zeros(run)
for runi in range(0,run):
    testdf=pd.read_csv("./data/testdata/testdata" + '_{}.csv'.format(runi+1)) 
    test_data_QRGMM=pd.read_csv("./data/testdata_GMM/testdata_GMM" + '_{}.csv'.format(runi+1))
    testdf_generated=pd.read_csv("./data/testdata_CWGAN/testdata_CWGAN" + '_{}.csv'.format(runi+1))
    
    WD_QRGMM[runi]=wasserstein_distance(test_data_QRGMM.iloc[:,0],testdf.iloc[:,3])
    WD_CWGAN[runi]=wasserstein_distance(testdf_generated.iloc[:,3], testdf.iloc[:,3])
    
    
    KS_QRGMM[runi],_=stats.kstest(test_data_QRGMM.iloc[:,0],testdf.iloc[:,3])
    KS_CWGAN[runi],_=stats.kstest(testdf_generated.iloc[:,3], testdf.iloc[:,3])
```

```python
fig,a =  plt.subplots(1,2,figsize=(15, 6))
bins1=np.arange(0,4.4,0.12)
bins2=np.arange(0,0.23,0.006)
a[0].hist(WD_QRGMM, bins=bins1,density=False, color='blue',alpha=0.6,label='QRGMM',edgecolor='w')
a[0].hist(WD_CWGAN, bins=bins1,density=False, color='red',alpha=0.6,label='CWGAN',edgecolor='w')
a[0].legend(fontsize=18)
a[0].tick_params(labelsize=18)
a[0].set_title("Test Problem 1: WD",fontsize=18)
col_labels = ['Mean','Std']
row_labels = ['QRGMM','CWGAN']
table_vals = [["{:.4f}".format(np.mean(WD_QRGMM)),"{:.4f}".format(np.std(WD_QRGMM))],["{:.4f}".format(np.mean(WD_CWGAN)),"{:.4f}".format(np.std(WD_CWGAN))]]
row_colors = ['blue','red']
my_table = a[0].table(cellText=table_vals, colWidths=[0.1]*2,
                     rowLabels=row_labels, colLabels=col_labels,edges='closed',
                     loc='center right')
my_table.auto_set_font_size(False)
my_table.set_fontsize(18)
my_table.scale(2, 2)


a[1].hist(KS_QRGMM, bins=bins2,density=False, color='blue',alpha=0.6,label='QRGMM',edgecolor='w')
a[1].hist(KS_CWGAN, bins=bins2,density=False, color='red',alpha=0.6,label='CWGAN',edgecolor='w')
a[1].legend(fontsize=18)
a[1].tick_params(labelsize=18)
a[1].set_title("Test Problem 1: KS",fontsize=18)
col_labels = ['Mean','Std']
row_labels = ['QRGMM','CWGAN']
table_vals = [["{:.4f}".format(np.mean(KS_QRGMM)),"{:.4f}".format(np.std(KS_QRGMM))],["{:.4f}".format(np.mean(KS_CWGAN)),"{:.4f}".format(np.std(KS_CWGAN))]]
row_colors = ['blue','red']
my_table = a[1].table(cellText=table_vals, colWidths=[0.1]*2,
                     rowLabels=row_labels, colLabels=col_labels,edges='closed',
                     loc='center right')
my_table.auto_set_font_size(False)
my_table.set_fontsize(18)
my_table.scale(2, 2)
fig.savefig('./plot/WDKS_test_withtable_testproblem1.pdf',dpi=300)
plt.show()
```
