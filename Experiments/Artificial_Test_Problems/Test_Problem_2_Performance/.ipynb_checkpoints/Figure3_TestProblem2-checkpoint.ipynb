```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

### Plot for empirical conditional distribution at $x^*$

```python
x1lb=0;
x1ub=10;
x2lb=-5;
x2ub=5;
# range of covariates
```

```python
def laplacefun(x, mu, lambda_):  # pdf of the laplace distribution 
    pdf = (1/(2*lambda_)) * np.e**(-1*(np.abs(x-mu)/lambda_))
    return pdf

def g1fun(x1,x2): # mean function of Test Problem 2
    g1=0.05*x1*x2
    return g1 

def g2fun(x1,x2): # standard deviation function of Test Problem 2
    g2=(5*(np.sin(x1+x2)**2)+5)
    return g2 
```

```python
# covariates of test data, x0=(4,4)
testx1=4
testx2=4 
# data applied to plot underlying true conditional probability density function curves
xl=-60
xr=60
X_pdf = np.arange(xl,xr,0.1)
Y_pdf = laplacefun(X_pdf, g1fun(testx1,testx2),g2fun(testx1,testx2))  
```

```python
runi = 0
test_data_QRGMM_run0 = pd.read_csv(
    f"./data/ctestdata_QRGMM/ctestdata_QRGMM_{runi+1}.csv"
)
test_data_CWGAN_run0 = pd.read_csv(
    f"./data/ctestdata_CWGAN/ctestdata_CWGAN_{runi+1}.csv"
)
test_data_Diff_run0 = pd.read_csv(
    f"./data/ctestdata_Diffusion/ctestdata_Diffusion_{runi+1}.csv"
)
test_data_Rect_run0 = pd.read_csv(
    f"./data/ctestdata_RectFlow/ctestdata_RectFlow_{runi+1}.csv"
)

runi = 2
test_data_QRGMM_run1 = pd.read_csv(
    f"./data/ctestdata_QRGMM/ctestdata_QRGMM_{runi+1}.csv"
)
test_data_CWGAN_run1 = pd.read_csv(
    f"./data/ctestdata_CWGAN/ctestdata_CWGAN_{runi+1}.csv"
)
test_data_Diff_run1 = pd.read_csv(
    f"./data/ctestdata_Diffusion/ctestdata_Diffusion_{runi+1}.csv"
)
test_data_Rect_run1 = pd.read_csv(
    f"./data/ctestdata_RectFlow/ctestdata_RectFlow_{runi+1}.csv"
)
```

```python
plt.rcParams['hatch.linewidth'] = 1.5

# 2 rows (rep1, rep2) x 4 cols (models)
fig, ax = plt.subplots(2, 4, figsize=(22, 10), sharex=True, sharey=True)

# keep your TP2 bins setting
bins = np.arange(xl, xr, 3)

# ---- model style config (keep your current scheme) ----
models = [
    ("CWGAN",     "red",        "\\\\"),
    ("Diffusion", "tab:orange", "xx"),
    ("RectFlow",  "tab:purple", "///"),
    ("QRGMM",     "blue",       ".."),
]

# ---- data mapping for each rep (TP2 uses column "10") ----
rep_data = [
    {  # rep1
        "CWGAN": test_data_CWGAN_run0["10"],
        "Diffusion": test_data_Diff_run0["10"],
        "RectFlow": test_data_Rect_run0["10"],
        "QRGMM": test_data_QRGMM_run0,
    },
    {  # rep2
        "CWGAN": test_data_CWGAN_run1["10"],
        "Diffusion": test_data_Diff_run1["10"],
        "RectFlow": test_data_Rect_run1["10"],
        "QRGMM": test_data_QRGMM_run1,
    }
]

row_names = ["rep1", "rep2"]

for r in range(2):
    for c, (name, color, hatch) in enumerate(models):
        a = ax[r, c]

        # True curve
        a.plot(X_pdf, Y_pdf, color='black', linewidth=3, label='True', zorder=5)

        # One model hist
        a.hist(rep_data[r][name],
               bins=bins, density=True,
               color=color, alpha=0.6,
               label=name, edgecolor='w',
               hatch=hatch, zorder=3)

        a.set_xlim(xl, xr)
        a.set_ylim(0, 0.18)
        a.tick_params(labelsize=22)

        # Column titles (model names) on top row only
        if r == 0:
            a.set_title("Test Problem 2: " + name, fontsize=22)

        # Row labels on the leftmost column only
        if c == 0:
            a.set_ylabel(row_names[r], fontsize=22)

        # Small legend per subplot (only 2 entries)
        a.legend(fontsize=22, frameon=False, loc="upper right")


# ---------- save ----------
os.makedirs("./plot", exist_ok=True)
fig.savefig('./plot/testproblem2_x0_plot.pdf',
            bbox_inches="tight", dpi=300)
plt.show()
```
