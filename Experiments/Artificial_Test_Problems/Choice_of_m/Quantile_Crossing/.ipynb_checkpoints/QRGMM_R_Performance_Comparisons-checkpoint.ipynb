# QRGMM v.s. QRGMM-R

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy.stats import wasserstein_distance
from scipy import stats
```

### Online Time and Crossing Rate

```python
onlinetime = pd.read_csv("./data_QC/onlinetime.csv").to_numpy()
crossing_rate = pd.read_csv("./data_QC/crossing_rate.csv").to_numpy().flatten()
```

```python
# ============================ online time summary ============================

mean_QRGMM_R = np.mean(onlinetime[:, 0])
std_QRGMM_R  = np.std(onlinetime[:, 0], ddof=1)

mean_QRGMM   = np.mean(onlinetime[:, 1])
std_QRGMM    = np.std(onlinetime[:, 1], ddof=1)

print("========== Online Time Summary ==========")
print(f"QRGMM-R : mean = {mean_QRGMM_R:.4f} s, std = {std_QRGMM_R:.6f} s")
print(f"QRGMM   : mean = {mean_QRGMM:.4f} s, std = {std_QRGMM:.6f} s")
print("=========================================")
```

```python
# ============================ crossing rate summary ============================

mean_crossing = np.mean(crossing_rate)
std_crossing  = np.std(crossing_rate, ddof=1)

print("========== Crossing Rate Summary ==========")
print(f"Crossing rate : mean = {mean_crossing:.4f}, std = {std_crossing:.4f}")
print("==========================================")
```

### Sample Mean and Sample SD

```python
run = 100

# ---------- per-run statistics ----------
mean_test    = np.zeros(run)
mean_QRGMM   = np.zeros(run)
mean_QRGMM_R = np.zeros(run)

std_test     = np.zeros(run)
std_QRGMM    = np.zeros(run)
std_QRGMM_R  = np.zeros(run)

WD_QRGMM     = np.zeros(run)
WD_QRGMM_R   = np.zeros(run)

KS_QRGMM     = np.zeros(run)
KS_QRGMM_R   = np.zeros(run)

for runi in range(run):
    test_data = pd.read_csv(
        "./data_QC/test_data/test_data_{}.csv".format(runi + 1),
        header=None
    )
    test_data_QRGMM = pd.read_csv(
        "./data_QC/test_data_QRGMM/test_data_QRGMM_{}.csv".format(runi + 1)
    )
    test_data_QRGMM_R = pd.read_csv(
        "./data_QC/test_data_QRGMM_R/test_data_QRGMM_R_{}.csv".format(runi + 1)
    )

    # ---------- mean ----------
    mean_test[runi]    = test_data.iloc[:, 10].mean()
    mean_QRGMM[runi]   = test_data_QRGMM.iloc[:, 0].mean()
    mean_QRGMM_R[runi] = test_data_QRGMM_R.iloc[:, 0].mean()

    # ---------- std (sample std) ----------
    std_test[runi]     = test_data.iloc[:, 10].std(ddof=1)
    std_QRGMM[runi]    = test_data_QRGMM.iloc[:, 0].std(ddof=1)
    std_QRGMM_R[runi]  = test_data_QRGMM_R.iloc[:, 0].std(ddof=1)

    # ---------- WD ----------
    WD_QRGMM[runi] = wasserstein_distance(
        test_data_QRGMM.iloc[:, 0],
        test_data.iloc[:, 10]
    )
    WD_QRGMM_R[runi] = wasserstein_distance(
        test_data_QRGMM_R.iloc[:, 0],
        test_data.iloc[:, 10]
    )

    # ---------- KS (two-sample, correct) ----------
    KS_QRGMM[runi], _ = stats.ks_2samp(
        test_data_QRGMM.iloc[:, 0],
        test_data.iloc[:, 10]
    )
    KS_QRGMM_R[runi], _ = stats.ks_2samp(
        test_data_QRGMM_R.iloc[:, 0],
        test_data.iloc[:, 10]
    )
```

```python
summary = pd.DataFrame({
    "mean(mean)": [
        mean_test.mean(),
        mean_QRGMM.mean(),
        mean_QRGMM_R.mean()
    ],
    "std(mean)": [
        mean_test.std(ddof=1),
        mean_QRGMM.std(ddof=1),
        mean_QRGMM_R.std(ddof=1)
    ],
    "mean(std)": [
        std_test.mean(),
        std_QRGMM.mean(),
        std_QRGMM_R.mean()
    ],
    "std(std)": [
        std_test.std(ddof=1),
        std_QRGMM.std(ddof=1),
        std_QRGMM_R.std(ddof=1)
    ]
}, index=["Truth", "QRGMM", "QRGMM-R"])

print("\nSummary statistics:\n")
print(summary.round(6))
```

### Plot for empirical conditional distribution at $x^*$

```python
runi=0
test_data_QRGMM_run0 = pd.read_csv("./data_QC/test_data_QRGMM/test_data_QRGMM" + '_{}.csv'.format(runi+1))
test_data_QRGMM_R_run0  = pd.read_csv("./data_QC/test_data_QRGMM_R/test_data_QRGMM_R" + '_{}.csv'.format(runi+1))
```

```python
xl=-60
xr=60

testx1=4
testx2=4

def laplacefun(x, mu, lambda_):  # pdf of the laplace distribution 
    pdf = (1/(2*lambda_)) * np.e**(-1*(np.abs(x-mu)/lambda_))
    return pdf

def g1fun(x1,x2): # mean function of Scenario 2
    g1=0.05*x1*x2
    return g1 

def g2fun(x1,x2): # standard deviation function of Scenario 2
    g2=(5*(np.sin(x1+x2)**2)+5)
    return g2 

X_pdf = np.arange(xl,xr,0.1)
Y_pdf = laplacefun(X_pdf, g1fun(testx1,testx2),g2fun(testx1,testx2))
```

```python
plt.rcParams['hatch.linewidth'] = 1.5

fig, a = plt.subplots(1, 2, figsize=(15, 6))
bins = np.arange(xl, xr, 2.5)

# ---- QRGMM ----
a[0].plot(X_pdf, Y_pdf, color='black', label='true', linewidth=3, linestyle="solid")
a[0].hist(test_data_QRGMM_run0, bins,
          density=True, color='blue', alpha=0.6,
          label='QRGMM', edgecolor='w',
          hatch='..')   
a[0].set_xlim(xl, xr)
a[0].set_ylim(0, 0.11)
a[0].tick_params(labelsize=20)
a[0].legend(fontsize=22)

# ---- QRGMM-R ----
a[1].plot(X_pdf, Y_pdf, color='black', label='true', linewidth=3, linestyle="solid")
a[1].hist(test_data_QRGMM_R_run0, bins,
          density=True, color='brown', alpha=0.4,
          label='QRGMM-R', edgecolor='w',
          hatch='++')   
a[1].set_xlim(xl, xr)
a[1].set_ylim(0, 0.11)
a[1].tick_params(labelsize=20)
a[1].legend(fontsize=22)

os.makedirs("./plot", exist_ok=True)
fig.savefig('./plot/QC_testproblem2_x0_plot.pdf', bbox_inches="tight", dpi=300)
plt.show()
```

### Plot for WD and KS

```python
plt.rcParams['hatch.linewidth'] = 1.5

fig, a = plt.subplots(1, 2, figsize=(15, 6))

bins1 = np.arange(1, 9, 0.18)
bins2 = np.arange(0.025, 0.25, 0.005)

# ============================ WD ============================

a[0].hist(WD_QRGMM, bins=bins1, density=False,
          color='blue', alpha=0.6,
          label='QRGMM', edgecolor='w',
          hatch='..')          

a[0].hist(WD_QRGMM_R, bins=bins1, density=False,
          color='brown', alpha=0.4,
          label='QRGMM-R', edgecolor='w',
          hatch='++')          

a[0].legend(fontsize=18)
a[0].tick_params(labelsize=18)
a[0].set_title("WD", fontsize=18)

col_labels = ['Mean', 'Std']
row_labels = ['QRGMM', 'QRGMM-R']
table_vals = [
    ["{:.6f}".format(np.mean(WD_QRGMM)),
     "{:.6f}".format(np.std(WD_QRGMM, ddof=1))],
    ["{:.6f}".format(np.mean(WD_QRGMM_R)),
     "{:.6f}".format(np.std(WD_QRGMM_R, ddof=1))]
]

my_table = a[0].table(
    cellText=table_vals,
    colWidths=[0.15] * 2,
    rowLabels=row_labels,
    colLabels=col_labels,
    edges='closed',
    loc='center right'
)
my_table.auto_set_font_size(False)
my_table.set_fontsize(15)
my_table.scale(1.5, 1.5)

# ============================ KS ============================

a[1].hist(KS_QRGMM, bins=bins2, density=False,
          color='blue', alpha=0.6,
          label='QRGMM', edgecolor='w',
          hatch='..')          

a[1].hist(KS_QRGMM_R, bins=bins2, density=False,
          color='brown', alpha=0.4,
          label='QRGMM-R', edgecolor='w',
          hatch='++')

a[1].legend(fontsize=18)
a[1].tick_params(labelsize=18)
a[1].set_title("KS", fontsize=18)

table_vals = [
    ["{:.6f}".format(np.mean(KS_QRGMM)),
     "{:.6f}".format(np.std(KS_QRGMM, ddof=1))],
    ["{:.6f}".format(np.mean(KS_QRGMM_R)),
     "{:.6f}".format(np.std(KS_QRGMM_R, ddof=1))]
]

my_table = a[1].table(
    cellText=table_vals,
    colWidths=[0.15] * 2,
    rowLabels=row_labels,
    colLabels=col_labels,
    edges='closed',
    loc='center right'
)
my_table.auto_set_font_size(False)
my_table.set_fontsize(15)
my_table.scale(1.5, 1.5)

fig.savefig('./plot/QC_WDKS_test_withtable_testproblem2.png', dpi=300)
plt.show()
```
